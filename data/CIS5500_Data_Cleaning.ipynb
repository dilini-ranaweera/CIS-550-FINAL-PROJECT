First we start by examining the Airbnb data. According to the schema of our database, the attributes of an Airbnb listing that can't be null are id and price. Examination of the data shows that there are no entries that have null values for id or price so no entries had to be removed.

```python
#Add Year column to both datasets.

ab_2023["year"] = pd.to_datetime('2023').year
ab_2020["year"] = pd.to_datetime('2020').year
ab_2020
```

```python
import numpy as np

def random_neighborhood(neigh_list_str):
    neigh_list = neigh_list_str.split(",")
    return np.random.choice(neigh_list)

# Apply the function to each row in the 'neighborhood' column
ab_2020['neighbourhood'] = ab_2020['neighbourhood'].apply(random_neighborhood)
ab_2020
```

```python
ab_df = pd.concat([ab_2020, ab_2023], axis =0)
```

```python
ab_df
```

```python
ab_df.dropna(subset=["id", "price", "city"], inplace=True)
```

```python
ab_df.info()
```

```python
ids = [i for i in range(1, len(ab_df) +1 )]
```

```python
ab_df["id"] = ids
```

```python
ab_df = ab_df.rename(columns={'neighbourhood': 'neighborhood'})
```

```python
ab_listings = ab_df[["id", "name", "price", "neighborhood", "city"]]
ab_listings
```

```python
ab_listings["neighborhood"].unique()
```

```python
ab_subclass = ab_df[["id", "minimum_nights", "room_type", "last_review"]]
ab_subclass["last_review"] = pd.to_datetime(ab_subclass["last_review"])
ab_subclass
```
